```python
# Hàm tách farme từ video
def extract_frames_from_video(video_path, output_folder, label_prefix, video_index, frame_interval_seconds=1):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)

    if fps == 0 or not cap.isOpened():
        print(f"Không mở được video: {video_path}")
        return 0

    interval = int(fps * frame_interval_seconds)
    frame_count = 0
    saved_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % interval == 0:
            filename = f"{label_prefix}_{video_index:03d}_f{saved_count}.jpg"
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, frame)
            saved_count += 1
        frame_count += 1

    cap.release()
    return saved_count
```

```python
def extract_frames_batch(video_folder, output_folder, label_prefix, frame_interval_seconds=1.5):
    os.makedirs(output_folder, exist_ok=True)
    video_files = sorted(os.listdir(video_folder))
    total_saved = 0

    print(f"\nĐang trích frame từ video {label_prefix.upper()}...")

    for idx, fname in enumerate(video_files):
        video_path = os.path.join(video_folder, fname)
        print(f"Đang xử lý: {fname} ...", end="")
        saved = extract_frames_from_video(video_path, output_folder, label_prefix, idx, frame_interval_seconds)
        total_saved += saved
        print(f"Đã lưu {saved} frame.")

    print(f"\nĐã hoàn tất. Tổng số frame {label_prefix.upper()} đã lưu: {total_saved}")
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
```

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
```

```python
for (x, y, w, h) in faces:
    face_img = cv2.rectangle(img, (x,y), (x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew,ey+eh), (0,255,0), 2)
        
plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```

**Từ các bước trên, có thể có được một với nhiệm vụ sau:**
* Đọc file ảnh -> chuyển ảnh thành dạng "gray"
* Sử dụng face_casecade để phát hiện gương mặt
* Sử dụng eye_casecade để phát hiện vùng mắt
* Trả về ảnh nếu số lượng vùng mắt >=2

```python
def get_cropped_image_if_2_eyes(image_path, is_fake=True):
    img = cv2.imread(image_path)
    if img is None:
        return None
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Với ảnh FAKE: chỉ chấp nhận nếu có đúng 1 gương mặt
    # if is_fake and len(faces) != 1:
    #     return None

    # Với ảnh REAL: vẫn duyệt từng mặt (nếu có >1)
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)

        if len(eyes) >= 2:
            return roi_color  # trả về khuôn mặt hợp lệ đầu tiên

    return None  # không có gương mặt nào hợp lệ
```

### Crop gương mặt từ các frame sau đó lưu vào folder

```python
# Lọc ra những ảnh bị mờ
def is_blurry(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
    return laplacian_var < 25  # ngưỡng độ mờ
```

```python
def process_and_save_faces(src_folder, dst_folder, label_prefix, is_fake=True):
    os.makedirs(dst_folder, exist_ok=True)
    saved_count = 0
    print("Đang cắt ảnh " + label_prefix +"...")
    for idx, fname in enumerate(sorted(os.listdir(src_folder))):
        img_path = os.path.join(src_folder, fname)
        cropped_face = get_cropped_image_if_2_eyes(img_path, is_fake=is_fake)

        if cropped_face is not None and not is_blurry(cropped_face):
            face_resized = cv2.resize(cropped_face, (128, 128))
            save_path = os.path.join(dst_folder, f"{label_prefix}_{saved_count:04d}.jpg")
            cv2.imwrite(save_path, face_resized)
            saved_count += 1
    print(f"\nTổng số ảnh {label_prefix} đã lưu: {saved_count}")
```
